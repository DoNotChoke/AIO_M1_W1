```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import math
```

```python
## Cau 1
def check_input(tp, fp, fn):
    if not (str)(tp).isnumeric():
        print("tp must be int")
        return False;
    if not (str)(fp).isnumeric():
        print("fp must be int")
        return False;
    if not (str)(fn).isnumeric():
        print("fn must be int")
        return False;
    if(tp <= 0 or fp <= 0 or fn <= 0):
        print("tp and fp and fn must be greater than zero")
        return False;
    return True;

def evaluate_classification_model(tp, fp, fn):
    if not check_input(tp, fp, fn):
        return;
    precision= tp/ (tp+fp)
    recall= tp/ (tp+fn)
    f1_score=2*(precision* recall)/(precision+ recall)
    print(f"precision is {precision}")
    print(f"recall is {recall}")
    print(f"f1_score is {f1_score}")
evaluate_classification_model(tp=2, fp='a', fn= 4)
evaluate_classification_model(tp=2, fp=3, fn='a')
evaluate_classification_model(tp=2, fp=3, fn=0)
```

## Relu function

```python
def Relu_func(x):
    return max(x,0.0)
    
x= [i for i in range (-20,20)]
y=[]
for i in x:
    y.append(Relu_func(i))
print(y)
plt.xlabel("x")
plt.ylabel("Relu(x)")
plt.plot(x,y)
```

## ELU function

```python
def ELU_func(x):
    alpha= 0.01
    if(x >0):
        return x;
    else:
        return alpha*(math.e**x-1)
x= -1
print(ELU_func(x))
x=[i for i in range (-20,20)]
y=[]
for i in x:
    y.append(ELU_func(i))
plt.xlabel("x")
plt.ylabel("ELU(x)")
plt.plot(x,y)
```

```python
## Cau 2
def is_number(n):
    try:
        float(n)
    except ValueError:
        return False;
    return True;

def check_string(n):
    if(n== "sigmoid" or n=="relu" or n=="elu"):
        return True
    return False
```

```python
def activation_func():
    x= input("Input x =")
    if not is_number(x):
        print("x must be a number")
        return;
    activation_function= input("Input activation Function (sigmoid|relu|elu):")
    if not check_string(activation_function):
        print(activation_function+" is not supported")
        return;
    x= (float)(x)
    if(activation_function== "sigmoid"):
        print(activation_function+ f" f({x}) = {sigmoid_func(x)}")
    elif(activation_function=="relu"): 
        print(activation_function+ f" f({x}) = {Relu_func(x)}")
    else: 
        print(activation_function+ f" f({x}) = {ELU_func(x)}")


activation_func()
```

```python
import random
def regression_loss_func():
    n= input("Input number of samples (interger number) which are generated: ")
    if not n.isnumeric():
        print("number of samples must be an interger number")
        return
    n= (int)(n)
    loss_name= input("Input loss name: ")
    loss=[]
    for i in range (n):
        predict= random.uniform(0,10)
        target= random.uniform(0,10)
        print(f"loss name: {loss_name}, sample: {i}, pred: {predict}, target: {target},")
        lo=0
        if(loss_name=="MAE"):
            lo= abs(target- predict)
        else:
            lo= (target- predict)**2
        print(f"loss: {lo}")
        loss.append(lo)
        
    if(loss_name=="MAE" or loss_name=="MSE"):
        print(f"final {loss_name}: {sum(loss)/n}")
    else:
        print(f"final {loss_name}: {math.sqrt(sum(loss))/n}")

regression_loss_func()
```

```python
def factorial(n):
    ans=1
    for i in range (1,n+1):
        ans = ans * i
    return ans;

def approx_sin(x, n):
    ans=0
    for i in range (n+1):
        ans= ans+ ((-1)**i)*((x**(2*i+1))/ factorial(2*i+1))
    return ans;
    
def approx_cos(x, n):
    ans=0
    for i in range (n+1):
        ans= ans+ ((-1)**i)*((x**(2*i))/ factorial(2*i))
    return ans;
    
def approx_sinh(x, n):
    ans= 0
    for i in range(n+1):
        ans= ans+ (x**(2*i+1))/factorial(2*i+1)
    return ans;
    
def approx_cosh(x,n):
    ans=0
    for i in range(n+1):
        ans= ans+ (x**(2*i))/ factorial(2*i)
    return ans;

x= (float)(input())
n= (int)(input())
print(approx_sin(x,n))
print(approx_cos(x,n))
print(approx_sinh(x,n))
print(approx_cosh(x,n))
```
